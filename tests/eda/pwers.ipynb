```python
from foe_foundry_data.powers import Powers
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display_markdown

powers = [p for _, p in Powers.PowerLookup.items()]
lengths = np.array([len(p.feature_descriptions) for p in powers])
plt.hist(lengths, bins=20)
plt.xlabel("Number of Feature Descriptions")
plt.ylabel("Frequency")
plt.title("Distribution of Feature Descriptions in Powers")
plt.show()


big_descriptions = lengths > 1000
i = np.argmax(big_descriptions)
p = powers[i]
display_markdown(p.feature_descriptions, raw=True)






```
